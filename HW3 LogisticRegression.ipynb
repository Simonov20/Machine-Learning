```python
# если присутствуют не все данные, то можно использовать удаление данных, код ниже, при условии небольших отсуствий. Если данные отсутсвуеют в значительном кол-ве , то можно использовать Imputer.
#data = data[ pd.isnull( data['наименование столбца '] ) == 0 ]
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
#выбираем признаки, по которым будем предугадывать заработок
selectedColumns = data [ ['age', 'workclass', 'education',  'educational-num', 
                          'marital-status',  'relationship', 'race', 'occupation',
                          'gender', 'capital-gain', 'capital-loss', 'hours-per-week',
                          'native-country', 'income'] ]
```

```python
# преобразуем столбец с наименованиями в матрицу, которая будет состоять из значений
X = pd.get_dummies( selectedColumns, columns = [ 'workclass', 'education', 'marital-status', 'occupation',
                                                'relationship','race',  'gender', 'native-country'])

# удаляем из матрицы Х интересующий нас столбец с доходами "income"
del X['income']
X.head()
```

### Проведем нормировку значений

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler=StandardScaler().fit(X)
X_scaled = scaler.transform(X)
X_scaled = pd.DataFrame(X_scaled, columns=X.columns)
X_scaled.head()
```

### Переведем столбец с доходами с числовые значения

```python

from sklearn.preprocessing import LabelEncoder
le = LabelEncoder ()
```

```python
le.fit (data ['income'])
```

```python
le.classes_
```

```python
# пример "расшировки" столбца 

le.transform(['<=50K', '>50K', '<=50K'])
```

```python
# трансформируем столбец с доходами
y = pd.Series (data = le.transform (data['income']))
y.head()
```

### Строим модель логистической регрессии

```python
model = LogisticRegression()
```

```python
model.fit(X_scaled,y)
y_pred = model.predict(X_scaled)
```

### Используем местрику правильности предсказаний accuracy=(кол-во правильных предсказаний)/(кол-во target'ов)

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy_score (y, y_pred)
```
